```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns


```

```python
# Load the Excel file
file_path = '/content/drive/MyDrive/Colab Notebooks/Tesis Maestria /ETL/Resumen.xlsx'
data = pd.read_excel(file_path)

# Display the first few rows of the dataframe to understand its structure
data.head(5)
```

```python
# Descripción general de los datos
data.info()
```

```python
data.describe(include='all')
```

```python
# Análisis de valores faltantes
missing_values = data.isnull().sum()
missing_percentage = (missing_values / len(data)) * 100
missing_data = pd.DataFrame({'Valores Faltantes': missing_values, 'Porcentaje (%)': missing_percentage})
missing_data
```

```python
# Estadísticas descriptivas de las variables numéricas
data.describe()
```

```python
# Seleccionar solo las columnas numéricas para el análisis de correlaciones
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns
numeric_data = data[numeric_columns]

# Recalcular la matriz de correlaciones con las columnas numéricas
correlation_matrix = numeric_data.corr()

# Visualización de la matriz de correlaciones
plt.figure(figsize=(12,8))
heatmap = sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Matriz de Correlaciones')
plt.show()
```
